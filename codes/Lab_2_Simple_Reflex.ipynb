```python
import random as r

class environment:
    def __init__(self, size):
        self.size = size
        self.field = [['#']*(size+2)] + [['#']+[' ']*size+['#']]*size +[['#']*(size+2)]
        l = ['#', ' ']
        for i in range(1,51):
            self.field[i] = ['#'] +  r.choices(l, weights=[1,10], k = self.size) + ['#']
        
        
    def display(self):
        for i in range(self.size+2):
            for j in range(self.size+2):
                print(self.field[i][j], end='')
            print()
```

```python
env = environment(50)
```

```python
env.display()
```

```python
class agent:
    def __init__(self,env):
        self.env = env
        self.coins = 0
        self.distance = 0
        self.steps = 0
        self.x = 1
        self.y = 1
        
    def decide(self, x,y):
        if self.env.field[x+1][y] != '#':
            x = x+1
        elif self.env.field[x][y+1] != '#':
            y = y+1
        elif self.env.field[x-1][y] != '#':
            x = x-1
        elif self.env.field[x][y-1] == ' ' or self.env.field[x][y-1] == 'o':
            y = y-1
        else:
            return -1,-1
        return x,y
    
    def move(self, x,y):
        self.x = x
        self.y = y
        i=0
        while(i<100):
            [a,b] = self.decide(x,y)
            if [a,b] == [-1,-1]:
                print("Collision")
                break
            self.env.field[x][y] = '-'
            x,y = a,b
            print(a,b)
            i=i+1
```
